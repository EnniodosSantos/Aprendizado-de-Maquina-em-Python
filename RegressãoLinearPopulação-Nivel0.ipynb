# Regressão Linear Simples
- Prever um valor contínuo dentro de um dominio

## Problema
- Prevusão da população brasileira em qualquer período entre 1872 e 2010

```python
import pandas as pd

```

### Carregamento dos dados

```python
dados = pd.read_csv("https://raw.githubusercontent.com/ect-info/ml/master/dados/pop_brasil.csv")
```

### Treinamento do modelo

```python
#Variável independente
X = dados.iloc[:, [0] ].values
y = dados.iloc[:, [1] ].values
```

#### Ajuste

```python
from sklearn import linear_model

regressao = linear_model.LinearRegression()
```

```python
regressao.fit(X,y)
```

### Teste

```python
X_teste = [ [1895], [1923], [1945], [1978], [2020]]
y_previsto = regressao.predict (X_teste)
```

## Visualização

```python
import matplotlib.pyplot as plt

plt.scatter(X,y, color = "blue")
```

```python
plt.scatter(X,y, color = "blue")
plt.plot(X_teste, y_previsto, color = "red")

plt.show()
```
